```python
import speech_recognition as sr
import datetime
from gtts import gTTS
import os
import numpy as np
import time
import pandas as pd


df = pd.read_csv('mobiles.csv')
df = df.drop(['Unnamed: 0','Model Number','Audio Formats','Video Formats','Supported Networks','Network Type',
             'Memory Card Slot Type','Processor Type','Processor Core','Primary Clock Speed',
             'Width','Height','Weight','Graphics PPI'],axis=1)

final_list = []
for i in range(len(df)):
    temp_list = []
    flag = 1
    for j in range(len(df.loc[i])):
        if pd.isna(df.loc[i][j]):
            flag = 0
    try:  
        if flag == 1:
            temp_list.append(df.iloc[i][0].replace(' ',''))
            temp_list.append(df.iloc[i][1].replace(' ',''))
            temp_list.append(df.iloc[i][2].replace(' ',''))
            temp_list.append(df.iloc[i][3][:-7].replace(' ',''))
            temp_list.append(df.iloc[i][4].replace(' ',''))
            temp_list.append(df.iloc[i][5].replace(' ',''))
            temp_list.append(df.iloc[i][6].replace(' ',''))
            temp_list.append(df.iloc[i][7].replace(' ',''))
            temp_list.append(df.iloc[i][8][:-12].replace(' ',''))
            temp_list.append(df.iloc[i][9][:-13].replace(' ',''))
            temp_list.append(df.iloc[i][10].replace(' ',''))
            final_list.append(temp_list)
    
    except:
        continue
        
final_df = pd.DataFrame(final_list)
final_df.columns = df.columns

for key,val in enumerate(final_df['Color']):
    if 'Blue' in val:
        final_df['Color'][key] = 'Blue'
    elif 'Black' in val:
        final_df['Color'][key] = 'Black'
    elif 'Green' in val:
        final_df['Color'][key] = 'Green'
    elif 'Grey' in val:
        final_df['Color'][key] = 'Grey'
    elif 'Gray' in val:
        final_df['Color'][key] = 'Grey'
    elif 'White' in val:
        final_df['Color'][key] = 'White'
    elif 'Red' in val:
        final_df['Color'][key] = 'Red'
    elif 'Cyan' in val:
        final_df['Color'][key] = 'Cyan'
    elif 'Silver' in val:
        final_df['Color'][key] = 'Silver'
    elif 'Cream' in val:
        final_df['Color'][key] = 'Cream'
    elif 'Snow' in val:
        final_df['Color'][key] = 'Snow'
    elif 'Violet' in val:
        final_df['Color'][key] = 'Violet'
    elif 'Pink' in val:
        final_df['Color'][key] = 'Pink'
    elif 'Purple' in val:
        final_df['Color'][key] = 'Purple'
    elif 'Gold' in val:
        final_df['Color'][key] = 'Gold'
    elif 'Aqua' in val:
        final_df['Color'][key] = 'Aqua'
    elif 'Diamond' in val:
        final_df['Color'][key] = 'Diamond'
    elif 'Bronze' in val:
        final_df['Color'][key] = 'Bronze'
    else:
        final_df['Color'][key] = final_df['Color'][key]

color_list = list(final_df['Color'].value_counts()[:8].index)
display_size_list = list(final_df['Display Size'].value_counts()[:8].index)
resolution_list = list(final_df['Resolution'].value_counts()[:8].index)
os_list = list(final_df['Operating System'].value_counts()[:8].index)
internal_storage_list = list(final_df['Internal Storage'].value_counts()[:8].index)
ram_list = list(final_df['RAM'].value_counts()[:8].index)
expandable_storage_list = list(final_df['Expandable Storage'].value_counts()[:8].index)
primary_camera_list = list(final_df['Primary Camera'].value_counts()[:8].index)
secondary_camera_list = list(final_df['Secondary Camera'].value_counts()[:8].index)
battery_list = list(final_df['Battery Capacity'].value_counts()[:8].index)

class ChatBot():
    def __init__(self, name):
        print("----- starting up", name, "-----")
        self.name = name
        
    def wake_up(self, text):
        return True if self.name in text.lower() else False

    def speech_to_text(self):
        recognizer = sr.Recognizer()
        with sr.Microphone() as mic:
             print("Listening...")
             audio = recognizer.listen(mic)
        try:
             self.text = recognizer.recognize_google(audio)
             print("me --> ", self.text)
        except:
             print("me -->  ERROR")
                
    @staticmethod
    def text_to_speech(text):
        print("AI --> ", text)
        speaker = gTTS(text=text, lang="en", slow=False)
        speaker.save("res.mp3")
        os.system("start res.mp3")  #if you have a macbook->afplay or for windows use->start
#       
        
    @staticmethod
    def action_time():
        return datetime.datetime.now().time().strftime('%H:%M')
#and run the script after adding the above function to the AI class
# Execute the AI
if __name__ == "__main__":
    
    ai = ChatBot(name="manav")
    df = final_df
    while True:
        res = ''
        ai.speech_to_text()
         ## waking up
        if ai.wake_up(ai.text) is True:
            res = "Hello I am Manav the AI, Do you want any help with selection of mobiles??"
            ai.text_to_speech(res)
            time.sleep(7)
            ai.speech_to_text()
            if ai.text.capitalize() in ['Yes','Yaa','Yess','Ya','Ye']:
                res = 'How much RAM do you want in your mobile??Tell No if u dont want to use this filter'
                ai.text_to_speech(res)
                time.sleep(7)
                ai.speech_to_text()
                while ai.text.upper() not in ram_list and ai.text.upper() not in ['NO','NOO','NAA','NONE']:
                    res = 'May be there is no phone with your filters or speak again properly.\
                        How much RAM do you want in your mobile??Tell No if u dont want to use this filter'
                    ai.text_to_speech(res)
                    time.sleep(24)
                    ai.speech_to_text()
                    
                if ai.text.upper() in ram_list:
                    df = df[df['RAM'] == ai.text[:-2]+'GB']
                    res = 'Ok you want ' + ai.text[:-2]+'GB ram mobiles? How much internal storage do u want?Tell No if u dont?'
                            
                else:
                    
                    res = 'Ok you don\'t want any filters in ram of mobiles? How much internal storage do u want?Tell \
                            No if u dont?'
                ai.text_to_speech(res)
                time.sleep(10)
                ai.speech_to_text()
                while ai.text[:-3]+'GB' not in internal_storage_list and ai.text.upper() not in ['NO','NOO','NAA','NONE']:
                    res = 'May be there is no phone with your filters or speak again properly.\
                    How much Internal Storage do you want in your mobile??Tell No if u dont want to use this filter'
                    ai.text_to_speech(res)
                    time.sleep(14)
                    ai.speech_to_text()
                        
                if ai.text[:-3]+'GB' in internal_storage_list:
                    df = df[df['Internal Storage'] == ai.text[:-3]+'GB']
                    res = 'Ok you want ' + ai.text[:-3]+'GB internal storage mobiles ? How much mAh battery capacity do u want?Tell No if u dont?'
                else:
                    
                    res = 'Ok you don\'t want any filters in internal storage of mobiles ? How much mAh battery capacity do u want?Tell No if u dont?'
                ai.text_to_speech(res)
                time.sleep(12)
                ai.speech_to_text()
                
                
                
                while ai.text + 'mAh' not in battery_list and ai.text.upper() not in ['NO','NOO','NAA','NONE']:
                    res = 'May be there is no phone with your filters or speak again properly.\
                    How much Battery Capacity do you want in your mobile??Tell No if u dont want to use this filter'
                    ai.text_to_speech(res)
                    time.sleep(14)
                    ai.speech_to_text()
                    
                if ai.text + 'mAh' in battery_list:
                    df = df[df['Battery Capacity'] == ai.text+'mAh']
                    res = 'Ok you want ' + ai.text + ' mAh battery mobiles. Tell thanks to have your results'
                else:
                    
                    res = 'Ok you don\'t want any filters in battery of mobiles. Tell thanks to have your results'
                ai.text_to_speech(res)
                time.sleep(18)
                ai.speech_to_text()
                
                if ai.text.capitalize() in ['Thank','Thanks','Thank You']:
                    res = np.random.choice(
                          ["you're welcome!","anytime!",
                           "no problem!","cool!",
                           "I'm here if you need me!","peace out!"])
                    res = res + '. Here are your results'
                    ai.text_to_speech(res)
                    time.sleep(5)
                    break
        
    os.remove("res.mp3")
    
```
